# Exercício 1

Escreva uma função que receba uma lista com valores de receita obtidos em x dias e calcule a média móvel de três dias ao longo dos valores.

**Entrada**: Lista contendo valores de receita

**Saída**: Lista com média móvel

Exemplo:

Entrada: [1, 4, 6, 8, 10, 11, 15]

Operação:
* $\frac{1+4+6}{3}$
* $\frac{4+6+8}{3}$

Saída: [3.66, 6, 8, 9.66, 12]

```python
def gerar_dados_ma(size:int) -> list:
  """Gera dados do exercicio 1 e 2

  :param size: Número de observações
  :type size: int

  :return:
    lista: Lista com as observações
  """

  import random
  random.seed(15)

  if size <= 2:
    raise Exception('Tamanho deve ser no minimo 3')

  output_list = []
  output_list.append(random.gauss(0, 2))

  output_list.append(random.gauss(0, 2))
  for k in range(2, size):
    new_obs = 0.8*output_list[k-1] - 0.5*output_list[k-2] + random.gauss(0, 2)
    output_list.append(new_obs)

  return output_list
```

```python
def exercicio_1(lista)->list:
    media_movel = []
    for i in range(len(lista) - 2):
        soma = sum(lista[i:i+3])
        media = soma / 3
        media_movel.append(media)
    return media_movel

# Teste de Uso:
lista = gerar_dados_ma(5)
print(lista)
resultado = exercicio_1(lista)
print(resultado)
```

# Exercício 2
Escreva a função do exercício 1 porém com o parâmetro de dias da média móvel sendo um input.

**Entrada**: Dados, parâmetro da média móvel

**Saída**: Lista com as médias móveis

Observação: o código não deve funcionar caso o volume de dados seja menor que o dias necessários para a média móvel.

Exemplo:

Entrada: [1, 4, 6, 8, 10, 11, 15] e parâmetro = 4

Operação:
* $\frac{1+4+6+8}{4}$
* $\frac{4+6+8+10}{4}$

Saída: [4.75, 7, 8.75, 11]

```python
def exercicio_2(lista, dias)->list:
    media_movel = []
    limite = dias - 1
    for i in range(len(lista) - limite):
        soma = sum(lista[i:i+dias])
        media = soma / dias
        media_movel.append(media)
    return media_movel

# Teste de Uso:
lista = gerar_dados_ma(7)
print(lista)
dias = 5
resultado = exercicio_2(lista, dias)
print(resultado)
```

# Exercício 3
Desejamos realizar transformações em features a serem utilizadas em um modelo.

**Entrada**  receberemos uma lista com os valoers da feature

**Saída** uma lista com a feature; a feature utilizando a normalização z-score; feature utilizando a normalização maxmin


Z-score: $\frac{X - \bar{X}}{DP(X)}$.

$\bar{X}$ é média das features e $DP(X)$ é o desvio-padrão de x.

Normalização maxmin: $\frac{X-min(X)}{max(X)-min(X)}$

Exemplo:
Entrada: [1 , 2, 3]

Saída: [[1, 2, 3], [-1, 0, 1], [0, 0.5, 1]]

```python
def gerar_features(size: int) -> list:
  """Gera dados do exercicio 3

  :param size: Número de observações
  :type size: int
  :param features: Número de features
  :type features: int

  :return:
    lista: Dicionario com as observações
  """
  import random
  random.seed(15)

  if size <= 2:
    raise Exception('Tamanho deve ser no minimo 3')

  output_list = []
  for k in range(size):
    output_list.append(random.gauss(4, 3))

  return output_list
```

```python
def calc_maxmin(lista)->list:
  max_valor = max(lista)
  #print("Valor Max: ", max_valor)

  min_valor = min(lista)
  #print("Valor Min: ", min_valor)

  range = max_valor -  min_valor
  #print("Range: ", range)

  maxmin = [(x - min_valor) / range for x in lista]
  #print("MaxMin: ", maxmin)

  return maxmin

def calc_zscores(lista)->list:
  media = sum(lista)/len(lista)
  #print("Média: ", media)

  soma_desvios_quadrados = sum((x - media) ** 2 for x in lista)
  #print("Soma Desvio Quadraticos: ", soma_desvios_quadrados)

  variancia = soma_desvios_quadrados / (len(lista) - 1)
  #print("Variancia: ", variancia)

  desvio_padrao = variancia ** 0.5
  #print("Desvio Padrão: ", desvio_padrao)

  z_scores = [(x - media) / desvio_padrao for x in lista]
  return z_scores

def exercicio_3(lista)->list:
  print("Lista: ", lista)

  #Calculando Z-Scores
  z_scores = calc_zscores(lista)
  print("Z-Scores: ", z_scores)

 #Calculando MaxMin
  maxmin = calc_maxmin(lista)
  print("MaxMin: ", maxmin)

  lista_resultado = [lista, z_scores, maxmin]
  return lista_resultado

### Dados exemplo do exercicio ###
#lista = [1, 2, 3]

# Teste de Uso:
lista = gerar_features(3)
exercicio_3(lista)
```

# Exercício 4
Desejamos realizar essa transformação em um dicionário de features, a transformação a ser utilizada sera um input na função. Se o input for z-score a normalização z-score deve ocorrer e caso seja max-min a normalização max-min deve ocorrer.

**Entrada** Dicionário com as features e tipo de padronização (string)

**Saída** Dicionário com a feature padronizada.

Exemplo:

Entrada:
{'Feature 1': [1, 2, 3],
'Feature 2': [1, 4, 10]}

Tipo de padronização: max-min

Saída:
{'Feature 1': [0, 0.5, 1],
'Feature 2': [0, 0.4, 1]}

```python
def gerar_features_dict(size: int, features: int) -> list:
  """Gera dados do exercicio 4

  :param size: Número de observações
  :type size: int
  :param features: Número de features
  :type features: int

  :return:
    lista: Dicionario com as observações
  """
  import random
  random.seed(15)

  if size <= 2:
    raise Exception('Tamanho deve ser no minimo 3')

  if features <= 1:
    raise Exception('Número de features deve ser no minimo 2')

  output_dict = {}
  mean = 3
  sd = 1

  for j in range(features):
    output_dict[f'feature_{j}'] = []
    for k in range(size):
      output_dict[f'feature_{j}'].append(random.gauss(mean, sd))
    mean += 0.5
    sd *= 1.1

  return output_dict
```

```python
import numpy as np

#Calcular z-score usando o numpy
def z_score(dicionario)->dict:
    dict_normalizado = {}

    for feature, valores in dicionario.items():
        # Converte a lista de valores em um array numpy
        valores_array = np.array(valores)

        # Calcula a média e o desvio padrão
        media = np.mean(valores_array)
        desvio = np.std(valores_array)

        # Aplica a normalização Z-score
        z_score = [(valor - media) / desvio for valor in valores]

        # Atualiza o dicionário
        dict_normalizado[feature] = z_score

    return dict_normalizado


#Calcular max_min usando o numpy
def max_min(dicionario)->dict:
    dict_normalizado = {}

    for feature, valores in dicionario.items():
        # Converte a lista de valores em um array numpy
        valores_array = np.array(valores)

        # Calcula o valor máximo e mínimo
        valor_max = np.max(valores_array)
        valor_min = np.min(valores_array)

        # Aplica a normalização Max-Min
        max_min = [(valor - valor_min) / (valor_max - valor_min) for valor in valores]

        # Atualiza o dicionário
        dict_normalizado[feature] = max_min

    return dict_normalizado

def exercicio_4(dicionario, tipo_padronizacao)->dict:
  if tipo_padronizacao == "z-score":
    return z_score(dicionario)
  elif tipo_padronizacao == "max-min":
    return max_min(dicionario)

### Dados exemplo do exercicio ###
#dicionario = {'Feature 1': [1, 2, 3], 'Feature 2': [1, 4, 10]}

# Teste de Uso:
dicionario = gerar_features_dict(4, 3)
tipo_padronizacao = "z-score" ### Utilizar as opções "z-score" ou "max-min" ###
exercicio_4(dicionario,tipo_padronizacao)
```

# Exercício 5
Possuímos uma lista com o id do usuário e uma outra lista com o valor das compras realizadas por ele. Considerar que o usuário no indice 0 fez a compra do índice 0.

Supondo que o usuário aparece mais de uma vez criar um dicionário com chave id e valor máximo das compras

**Entrada**: Lista com id e lista com as compras

**Saída**: Dicionário com chave id do usuário e valor o máximo das compras

Exemplo:

Entrada:

Lista_id = [0, 1, 2, 1, 0]

Lista_compras = [100, 150, 200, 75, 300]

Saída:
{0: 300,
1: 150,
2: 200}


```python
from typing import Tuple
def gerar_listas(size: int, users: int) -> Tuple[list, list]:
  """Gera dados do exercicio 4

  :param size: Número de observações
  :type size: int
  :param users: Número de usuários
  :type features: int

  :return:
    Tuple[list, list]: Lista com o identificador dos usuários e lista com o valor da compra.
  """
  import random
  random.seed(15)

  if size <= users * 4:
    raise Exception('Amostra deve ser 4 vezes maior que o número de usuarios')



  id_user = []
  receita = []

  for k in range(size):
    id_user.append(random.randint(0, users))
    receita.append(random.gammavariate(10, 200))


  return id_user, receita
```

```python
def exercicio_5(dicionario)->dict:
  # Ler a tupla
  lista_id, lista_compras = dicionario

  # Dicionario de IDs com os valores máximos
  max_compras_por_usuario = {}

  # Ler dados das listas de IDs de usuários e valores
  for i in range(len(lista_id)):
    id_usuario = lista_id[i]
    valor_compra = lista_compras[i]

    # Verificar se o ID de usuário já está no dicionário e atualizar o valor máximo
    if id_usuario in max_compras_por_usuario:
      if valor_compra > max_compras_por_usuario[id_usuario]:
        max_compras_por_usuario[id_usuario] = valor_compra
    else:
      max_compras_por_usuario[id_usuario] = valor_compra

  return max_compras_por_usuario

### Dados exemplo do exercicio ###
#lista_id = [0, 1, 2, 1, 0]
#lista_compras = [100, 150, 200, 75, 300]
#dicionario = (lista_id, lista_compras)

# Teste de Uso:
dicionario = gerar_listas(12, 2)
exercicio_5(dicionario)
```
